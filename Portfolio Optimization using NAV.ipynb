```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tickers = ['TSLA','AAPL','MSFT','XOM','AMD','C']

data = yf.download(tickers, period='10y', interval='1d')
data
```

```python
adj_close_df = data['Adj Close']
for i in tickers:
    adj_close_df[i+' %Returns'] = adj_close_df[i].pct_change()
adj_close_df.dropna(inplace=True)

adj_close_df
```

```python
five_years_data = adj_close_df[:1257]
five_years_data
```

```python
returns_df = five_years_data.iloc[:,-6:]
returns_df
```

```python
return_cols = list(returns_df.columns)
returns_list=[]

for i in return_cols:
    temp = returns_df[i].mean()
    temp2 = ((temp+1)**250-1)*100
    returns_list.append(temp2)
```

```python
returns_list
```

```python
risk_list=[]

for i in return_cols:
    temp = returns_df[i].std()*np.sqrt(250)*100
    risk_list.append(temp)
```

```python
risk_list
```

```python
risk_returns_df = pd.DataFrame(index=return_cols)
risk_returns_df['Returns%'] = returns_list
risk_returns_df['Risk%'] = risk_list
risk_returns_df['Risk Adj Return'] = risk_returns_df['Returns%']/risk_returns_df['Risk%']
risk_returns_df
```

```python
risk_returns_list = list(risk_returns_df['Risk Adj Return'])
risk_returns_list
```

```python
sum = np.sum(risk_returns_list)
```

```python
weights = risk_returns_list/sum
```

```python
weights
```

```python
weights.sum()
```

```python
risk_returns_df['Weights'] = weights
risk_returns_df
```

```python
last_five_years = adj_close_df[1258:]
last_five_years
```

```python
last_five_ret = last_five_years.iloc[:,-6:]
last_five_ret
```

```python
rebased_df = pd.DataFrame()

for i in return_cols:
    rebased_list = [100]
    for j in list(last_five_ret[i]):
        temp = rebased_list[-1] * (j+1)
        rebased_list.append(temp)
    rebased_df[i] = rebased_list
```

```python
rebased_df
```

```python
plt.figure(figsize=(20,9))
plt.plot(rebased_df)
plt.legend(return_cols)
plt.show()
```

```python
final_df = pd.DataFrame()
final_df['Portfolio NAV'] = rebased_df.dot(list(risk_returns_df['Weights']))
```

```python
final_df
```

```python
spy = yf.download('SPY', period='5y', interval='1d')
spy['Returns'] = spy['Adj Close'].pct_change()
spy.dropna(inplace=True)

spy
```

```python
rebased_spy=[100]

for i in list(spy['Returns']):
    temp = rebased_spy[-1] * (i+1)
    rebased_spy.append(temp)
```

```python
final_df['SPY NAV'] = rebased_spy
final_df
```

```python
plt.figure(figsize=(20,9))
plt.plot(final_df)
plt.legend(final_df.columns)
plt.show()
```
